```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('D:/Datasets/Housing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Boxplot of Price
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of House Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Distribution of area
plt.figure(figsize=(10, 6))
sns.histplot(df['area'], bins=30, kde=True)
plt.title('Distribution of House Areas')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
# Piechart of number of bedrooms
plt.figure(figsize=(10, 6))
plt.pie(df['bedrooms'].value_counts(), labels=df['bedrooms'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Number of Bedrooms')
plt.show()
```

```python
# Piechart of number of parking spaces
plt.figure(figsize=(10, 6))
plt.pie(
    df["parking"].value_counts(), labels=df["parking"].unique(), autopct="%1.1f%%"
)
plt.title("Distribution of Number of Parking Spaces")
plt.show()
```

```python
# Number of houses on main road or not
plt.figure(figsize=(10, 6))
sns.countplot(x="mainroad", data=df)
plt.title("Distribution of Number of Bedrooms")
plt.show()
```

```python
# number of houses with a guest room or not
plt.figure(figsize=(10, 6))
sns.countplot(x="guestroom", data=df)
plt.title("Distribution of Houses with or without a Guest Room")
plt.show()
```

```python
# Number of houses with or without a basement
plt.figure(figsize=(10, 6))
sns.countplot(x="basement", data=df)
plt.title("Distribution of Houses with or without a Basement")
plt.show()
```

```python
# Furnising status of each house
plt.figure(figsize=(10, 6))
sns.countplot(x='furnishingstatus', data=df)
plt.title('Distribution of Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Count')
plt.show()
```

```python
# Number of houses with howaterheating or not
plt.figure(figsize=(10,6))
sns.countplot(x='hotwaterheating', data=df)
plt.title('Distribution of Houses with or without Hot Water Heating')
plt.xlabel('Hot Water Heating')
plt.ylabel('Count')
plt.show()
```

```python
# Number of houses with air conditioning
plt.figure(figsize=(10, 6))
sns.countplot(x='airconditioning', data=df)
plt.title('Distribution of Houses with or without Air Conditioning')
plt.xlabel('Air Conditioning')
plt.ylabel('Count')
plt.show()
```
